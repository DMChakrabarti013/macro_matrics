```python
import numpy as np
import pandas as pd
import cvxpy as cp

from sklearn.preprocessing    import StandardScaler
from sklearn.model_selection  import StratifiedKFold
from sklearn.linear_model     import LogisticRegression
from sklearn.metrics          import zero_one_loss

# ─── ASSUME THESE DATAFRAMES ARE ALREADY IN YOUR NAMESPACE ─────────────────
# selected_irfs             # DataFrame: ['State'] + 25 IRF columns
# bayesian_coefficients     # DataFrame: ['State'] + 25 lag-1 VAR coef columns
# baseline_df_synthetic     # DataFrame: ['State','YearQuarter', <numeric covariates>]
# ml_scores_df              # DataFrame: ['State','decision_tree_scores','rf_scores','xgb_scores','gpr_scores']
# pre_treatment_Y           # DataFrame: index=State, columns=pre-treatment quarters of gdp_growth

# ─── 1) COLUMN LISTS ─────────────────────────────────────────────────────────
irf_cols  = [c for c in selected_irfs.columns        if c != 'State']
coef_cols = [c for c in bayesian_coefficients.columns if c != 'State']
ml_cols   = ['decision_tree_score','rf_score','xgb_score','gpr_score']
states    = selected_irfs['State'].values

# ─── 2) BUILD STATIC BASELINE FEATURES ───────────────────────────────────────
numeric = baseline_df_synthetic.drop(columns=['YearQuarter']) \
          .select_dtypes(include=[np.number])
numeric['State'] = baseline_df_synthetic['State']
baseline_static = numeric.groupby('State').mean()  # one row per state

# ─── 3) SCALE IRFs & COEFFS ──────────────────────────────────────────────────
IRF_mat  = StandardScaler().fit_transform(selected_irfs[irf_cols].values)     
COEF_mat = StandardScaler().fit_transform(bayesian_coefficients[coef_cols].values)

# ─── 4) IRF “RELEVANCE” WEIGHTS ─────────────────────────────────────────────
m = pd.merge(
    bayesian_coefficients[['State'] + coef_cols],
    selected_irfs[['State'] + irf_cols],
    on='State'
)
corr_mat = m[irf_cols + coef_cols].corr().loc[irf_cols, coef_cols]
irf_w    = corr_mat.abs().max(axis=1).fillna(0).values  # length=25

# ─── 5) SYNTHETIC CONTROL WEIGHTS HELPER ────────────────────────────────────
def synth_weights(delta_i, delta_d, Omega_i, Omega_d, lambda2):
    n = delta_d.shape[0]
    w = cp.Variable(n, nonneg=True)
    constraints = [cp.sum(w) == 1]
    loss = cp.sum_squares(delta_i - delta_d.T @ w) \
         + lambda2 * cp.sum_squares(Omega_i - Omega_d.T @ w)
    prob = cp.Problem(cp.Minimize(loss), constraints)
    prob.solve(solver=cp.SCS, verbose=False)
    return w.value if w.value is not None else np.zeros(n)

# ─── 6) CV‐ERROR FOR GROUP CLASSIFIER ────────────────────────────────────────
kf     = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
ml_mat = merged_df.set_index('State')[ml_cols].loc[states].values  # align to `states`

def compute_cv_error(G):
    errs = []
    for train, test in kf.split(ml_mat, G):
        y_tr, y_te = G[train], G[test]
        if len(np.unique(y_tr)) < 2:
            continue
        clf = LogisticRegression(solver='liblinear')
        clf.fit(ml_mat[train], y_tr)
        errs.append(zero_one_loss(y_te, clf.predict(ml_mat[test])))
    return np.mean(errs) if errs else 0.0

# ─── 7) GRID SEARCH HYPERPARAMS ──────────────────────────────────────────────
alpha_grid   = np.linspace(0,1,11)
lambda2_grid = [0,0.1,0.5,1,2,5,10,20,50,100]
c_grid       = np.linspace(0.1,0.9,9)
gamma_grid   = np.linspace(0,1,11)

best = {'loss': np.inf}

for alpha in alpha_grid:
    Δ   = alpha * IRF_mat + (1-alpha) * COEF_mat      # (n_states,25)
    Xw  = Δ * irf_w[None,:]                          # apply relevance weight

    for lam2 in lambda2_grid:
        comp_score = ml_mat.mean(axis=1)             # composite ML score

        for c in c_grid:
            thr = np.quantile(comp_score, c)
            G   = (comp_score > thr).astype(int)

            cv_err = compute_cv_error(G)

            pre_errs = []
            for i, state in enumerate(states):
                if G[i] != 1:
                    continue

                # raw donors
                raw_states = states[G==0]

                # intersect to ensure baseline_static & pre_treatment_Y have them
                ds = sorted(set(raw_states)
                           & set(baseline_static.index)
                           & set(pre_treatment_Y.index))
                if not ds:
                    continue

                donor_idx = [j for j,s in enumerate(states) if s in ds]

                Δi = Xw[i,:]                   # (25,)
                Δd = Xw[donor_idx,:]           # (n_donors,25)

                Ωi = baseline_static.loc[state].values       # (n_covs,)
                Ωd = baseline_static.loc[ds].values          # (n_donors,n_covs)

                w = synth_weights(Δi, Δd, Ωi, Ωd, lam2)
                if len(w) != len(donor_idx):
                    continue

                Yi     = pre_treatment_Y.loc[state].values       # (T0,)
                Yd_mat = pre_treatment_Y.loc[ds].values          # (n_donors,T0)
                Ysyn   = Yd_mat.T @ w                            # (T0,)

                pre_errs.append(np.mean((Yi - Ysyn)**2))

            pre_err = np.mean(pre_errs) if pre_errs else 0.0

            for gamma in gamma_grid:
                loss = gamma * pre_err + (1-gamma) * cv_err
                if loss < best['loss']:
                    best.update({
                        'loss':    loss,
                        'alpha':   alpha,
                        'lambda2': lam2,
                        'c':       c,
                        'gamma':   gamma,
                        'pre_err': pre_err,
                        'cv_err':  cv_err
                    })

print("🏆 Best hyperparameters:", best)
```

```python
import numpy as np
import pandas as pd
import cvxpy as cp

from sklearn.preprocessing    import StandardScaler
from sklearn.model_selection  import StratifiedKFold
from sklearn.linear_model     import LogisticRegression
from sklearn.metrics          import zero_one_loss

# ─── ASSUME THESE DATAFRAMES ARE ALREADY IN YOUR NAMESPACE ─────────────────
# selected_irfs             # DataFrame: ['State'] + 25 IRF cols
# bayesian_coefficients     # DataFrame: ['State'] + 25 lag-1 VAR coef cols
# baseline_df_synthetic     # DataFrame: ['State','YearQuarter', <numeric covariates>]
# merged_df                 # DataFrame: ['State','decision_tree_score','rf_score','xgb_score','gpr_score']
# pre_treatment_Y           # DataFrame: index=State, cols=pre-treatment quarters of gdp_growth

# ─── 1) COLUMN LISTS ─────────────────────────────────────────────────────────
irf_cols  = [c for c in selected_irfs.columns        if c != 'State']
coef_cols = [c for c in bayesian_coefficients.columns if c != 'State']
ml_cols   = ['decision_tree_score','rf_score','xgb_score','gpr_score']
states    = selected_irfs['State'].values

# ─── 2) BUILD STATIC BASELINE FEATURES ───────────────────────────────────────
numeric = (
    baseline_df_synthetic
      .drop(columns=['YearQuarter'])
      .select_dtypes(include=[np.number])
)
numeric['State'] = baseline_df_synthetic['State']
baseline_static = numeric.groupby('State').mean()

# ─── 3) SCALE IRFs & COEFS ──────────────────────────────────────────────────
IRF_mat  = StandardScaler().fit_transform(selected_irfs[irf_cols].values)
COEF_mat = StandardScaler().fit_transform(bayesian_coefficients[coef_cols].values)

# ─── 4) IRF “RELEVANCE” WEIGHTS ─────────────────────────────────────────────
m        = pd.merge(
    bayesian_coefficients[['State'] + coef_cols],
    selected_irfs[['State'] + irf_cols],
    on='State'
)
corr_mat = m[irf_cols + coef_cols].corr().loc[irf_cols, coef_cols]
irf_w    = corr_mat.abs().max(axis=1).fillna(0).values  # length = 25

# ─── 5) SYNTHETIC CONTROL WEIGHTS HELPER ────────────────────────────────────
def synth_weights(delta_i, delta_d, Omega_i, Omega_d, lambda2):
    n = delta_d.shape[0]
    w = cp.Variable(n, nonneg=True)
    constraints = [cp.sum(w) == 1]
    loss = cp.sum_squares(delta_i - delta_d.T @ w) \
         + lambda2 * cp.sum_squares(Omega_i - Omega_d.T @ w)
    prob = cp.Problem(cp.Minimize(loss), constraints)
    prob.solve(solver=cp.SCS, verbose=False)
    return w.value if w.value is not None else np.zeros(n)

# ─── 6) CV‐ERROR FOR GROUP CLASSIFIER ────────────────────────────────────────
kf     = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
ml_mat = merged_df.set_index('State')[ml_cols].loc[states].values

def compute_cv_error(G):
    errs = []
    for train, test in kf.split(ml_mat, G):
        y_tr, y_te = G[train], G[test]
        if len(np.unique(y_tr)) < 2:
            continue
        clf = LogisticRegression(solver='liblinear')
        clf.fit(ml_mat[train], y_tr)
        errs.append(zero_one_loss(y_te, clf.predict(ml_mat[test])))
    return np.mean(errs) if errs else 0.0

# ─── 7) FIXED HYPERPARAMETERS ───────────────────────────────────────────────
alpha    = 0.5
gamma    = 0.5

# form the weighted & scaled features
Δ        = alpha * IRF_mat + (1 - alpha) * COEF_mat   # (n_states, 25)
Xw       = Δ * irf_w[None, :]                         # apply relevance weight

# composite ML score & threshold at median
comp_score = ml_mat.mean(axis=1)
c_threshold = np.median(comp_score)
G          = (comp_score > c_threshold).astype(int)

# compute CV‐error once
cv_err = compute_cv_error(G)

# ─── 8) GRID‐SEARCH ONLY λ₂ ─────────────────────────────────────────────────
lambda2_grid = [0,0.1,0.5,1,2,5,10,20,50,100]
best = {'loss': np.inf}

for lam2 in lambda2_grid:
    pre_errs = []
    for i, state in enumerate(states):
        if G[i] != 1:
            continue

        raw_states = states[G == 0]
        donors     = sorted(
            set(raw_states)
            & set(baseline_static.index)
            & set(pre_treatment_Y.index)
        )
        if not donors:
            continue

        donor_idx = [j for j,s in enumerate(states) if s in donors]

        Δi = Xw[i, :]                       # (25,)
        Δd = Xw[donor_idx, :]               # (n_donors,25)

        Ωi = baseline_static.loc[state].values   # (n_covs,)
        Ωd = baseline_static.loc[donors].values  # (n_donors,n_covs)

        w  = synth_weights(Δi, Δd, Ωi, Ωd, lam2)
        if len(w) != len(donor_idx):
            continue

        Yi     = pre_treatment_Y.loc[state].values       # (T0,)
        Yd_mat = pre_treatment_Y.loc[donors].values      # (n_donors,T0)
        Ysyn   = Yd_mat.T @ w                            # (T0,)

        pre_errs.append(np.mean((Yi - Ysyn)**2))

    pre_err = np.mean(pre_errs) if pre_errs else 0.0
    loss    = gamma * pre_err + (1-gamma) * cv_err

    if loss < best['loss']:
        best.update({
            'loss':    loss,
            'lambda2': lam2,
            'pre_err': pre_err,
            'cv_err':  cv_err
        })

print("🏆 Best λ₂ with (α=0.5, c=median, γ=0.5):", best)
```
