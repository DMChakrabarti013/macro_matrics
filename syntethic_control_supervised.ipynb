```python
import pandas as pd

selected_irfs = pd.read_csv('selected_irfs.csv')
bayesian_coefficients = pd.read_csv('bayesian_var_vectorized_coefficients.csv')
```

```python
selected_irfs
```

```python
# Keep only the first 25 columns of bayesian_coefficients (drop extra columns if any)
bayesian_coefficients = bayesian_coefficients.iloc[:, :26]
bayesian_coefficients
```

```python
baseline_df_synthetic = pd.read_csv('baseline_data_synthetic.csv')
group_supervised = pd.read_csv('group_supervised.csv')
```

```python
# Identify the columns in each dataframe
irf_cols = [col for col in selected_irfs.columns if col.startswith('irf_')]
coef_cols = [col for col in bayesian_coefficients.columns if col.startswith('coef_')]

# Multiply and then sum the values for each state
irf_sum_scaled = (selected_irfs[irf_cols] * 0.5)
coef_sum_scaled = (bayesian_coefficients[coef_cols] * 0.5)

```

```python

combined_dynamics = irf_sum_scaled.values + coef_sum_scaled.values
combined_dynamics_df = pd.DataFrame(combined_dynamics, columns=irf_sum_scaled.columns)
combined_dynamics_df.insert(0, 'State', selected_irfs['State'])
new_names = ['combined_' + str(i) for i in range(1, 26)]
combined_dynamics_df.columns = ['State'] + new_names
combined_dynamics_df
```

```python
combined_dynamics_df.to_csv('combined_dynamics.csv', index=False)
```

```python
# Find the common states across the three dataframes
common_states = set(baseline_df_synthetic['State']) & set(group_supervised['State']) & set(combined_dynamics_df['State'])
print("Common states:", common_states)

```

```python
print("Number of common states:", len(common_states))
```

```python
# Filter the dataframes to only include these common states
baseline_df_synthetic_filtered = baseline_df_synthetic[baseline_df_synthetic['State'].isin(common_states)]
group_supervised_filtered = group_supervised[group_supervised['State'].isin(common_states)]
combined_dynamics_df_filtered = combined_dynamics_df[combined_dynamics_df['State'].isin(common_states)]

# Display the filtered dataframes
baseline_df_synthetic_filtered
group_supervised_filtered
combined_dynamics_df_filtered
```
